```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargar el archivo
data = pd.read_csv('cdmx_sin_nulos.csv', index_col="id")
```

```python
data.info()
```

```python
#Eliminamos la columna Unnamed 
data1 = data.drop(['Unnamed: 0'], axis = 1)
data1
```

```python
data1.info()
```

```python
valores_nulos = data1.isnull().sum()
valores_nulos
```

```python
cuantitativas = data1.iloc[:, [7,8,10,11,18,19,20,21,23,24,25,26,27,30,31,32,33,36,37,38,39,40,41,42,45,46,47,48]]
cualitativas = data1.iloc[:, [0,1,2,3,4,5,6,9,12,13,14,15,16,17,22,28,29,34,35,43,44]]
```

```python
cuantitativas.info()
```

```python
cualitativas.info()
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe 
fig = plt.figure(figsize = (15,8))
cuantitativas.plot(kind = 'box', vert = False)
plt.title("Valores atÃ­picos del dataframe")
plt.show() #Dibuja el df
```

**APLICAMOS DESVIACION ESTANDAR PARA ELIMINAR OUTLIERS DEL DATAFRAME**

```python
y = cuantitativas
Limite_Superior = y.mean() + 3*y.std()
Limite_Inferior = y.mean() - 3*y.std()
print("LIMITE SUPERIOR PERMITIDO\n", Limite_Superior)
print("LIMITE INFERIOR PERMITIDO\n", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos
data2 = cuantitativas [(y <= Limite_Superior) & (y >= Limite_Inferior)]
data2
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
#Remplazar valores atÃ­picos (nulos) del df con "mean"
#Realizamos una copia del df
data_clean = data2.copy()
data_clean = data_clean.fillna(round(data2.mean(), 1))
data_clean
```

```python
valores_nulos = data_clean.isnull().sum()
valores_nulos
```

```python
#Unimos el df cuantitativo limpio con el cualitativo
df_completo = pd.concat([cualitativas, data_clean], axis = 1)
df_completo
```

```python
df_completo.to_csv("Desviacion_Estandar_CDMX.csv")
```

**APLICAMOS PROCEDIMIENTO DE CUARTILES PARA ELIMINAR OUTLIERS DEL DATAFRAME**

```python
#Metodo aplicando cuartiles. Encuentro cuartiles 0.25 y 0.75
y = cuantitativas
porcentil25 = y.quantile(0.25) #Q1
porcentil75 = y.quantile(0.75) #Q3
iqr = porcentil75 - porcentil25

Limite_Superior_iqr = porcentil75 +1.5*iqr
Limite_Inferior_iqr = porcentil25 -1.5*iqr

print("LIMITE SUPERIOR PERMITIDO\n", Limite_Superior_iqr)
print("LIMITE INFERIOR PERMITIDO\n", Limite_Inferior_iqr)
```

```python
#Obtener datos limpios del df
data3_iqr = cuantitativas[(y <= Limite_Superior_iqr) & (y >= Limite_Inferior_iqr)]
data3_iqr
```

```python
valores_nulos = data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos
data4_iqr = data3_iqr.copy()
data4_iqr = data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
valores_nulos = data4_iqr.isnull().sum()
valores_nulos
```

```python
#Unimos el df cuantitativo limpio con el cualitativo
Datos_limpios = pd.concat([cualitativas, data4_iqr], axis = 1)
Datos_limpios
```

```python
valores_nulos = Datos_limpios.isnull().sum()
valores_nulos
```

```python
Datos_limpios.info()
```

```python
Datos_limpios.to_csv("Cuartiles_CDMX.csv")
```
